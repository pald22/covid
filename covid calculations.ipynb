```python
import pandas as pd # Data Frames Python Data Analysis Library
import matplotlib.pyplot as plt # Graphing
import pylab
import seaborn as sns
from matplotlib import pyplot
```

```python
#https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv
df = pd.read_csv('owid-covid-data2.csv')
df['date'] = pd.to_datetime(df['date']) # recast into dt object

# Basic Analytics
def show_analytics(data):
  print(data.head(10)) # Shows first 10 rows
  print(data.info()) # Shows basic DF info
  print(data.describe()) # Shows data re. numerical columns
  print(data.dtypes) # Returns data types
#
#show_analytics(df)
countries = set(df.iloc[:, 2])
countries = list(sorted(countries))
#len countries = 211
df = df.drop(['iso_code', 'total_deaths', 'new_deaths', 'total_cases_per_million', 'new_cases_per_million', 'total_deaths_per_million', 'new_deaths_per_million', 'total_tests', 'total_tests_per_thousand', 'new_tests_per_thousand', 'tests_units'], 1)
for country in countries:
    print(country)
    test = df.loc[df['location']==country]
    test = test[test.new_cases != 0]
    #test = test[test.new_tests != 0]
    test['newrate'] = test['new_cases']/test['total_cases']
    #print(test)
    test['newrate2'] = test['newrate'].rolling(window=10).mean()

    fig,ax = plt.subplots(figsize = (20,15))
    test.plot(x='date', y='newrate2',title = f'{country} new cases / total cases:',marker='.',markersize=10,ax=ax)

    plt.yscale('log')
    plt.ylabel('new cases / total cases, log scale')
    plt.savefig(f'{country}.png')
    plt.clf()
    plt.close()
```
